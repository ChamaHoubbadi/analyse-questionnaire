```python
import matplotlib.pyplot as plt
import pandas as pd

influence_data = {
    'Influence des avis': ['Très influent', 'Influent', 'Moyennement influent', 'Peu influent', 'Pas du tout influent'],
    'Nombre de réponses': [38, 30, 12, 5, 3]
}

df_influence = pd.DataFrame(influence_data)

plt.figure(figsize=(8, 6))
plt.bar(df_influence['Influence des avis'], df_influence['Nombre de réponses'], color='lightblue', alpha=0.8)
plt.title("Répartition de l'influence des avis en ligne sur les décisions d'achat")
plt.xlabel("Influence des avis")
plt.ylabel("Nombre de réponses")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Graphique 2
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('reponses_questionnaire.csv')

ameliorations = df['Quelles améliorations espérez-vous de la part des entreprises utilisant l\'analyse des sentiments pour analyser les avis en ligne ?']

ameliorations_split = ameliorations.str.split(',').explode().str.strip()  
counts = ameliorations_split.value_counts()

plt.figure(figsize=(8, 6))
counts.plot(kind='bar', color='skyblue', alpha=0.7)
plt.title('Répartition des catégories d\'améliorations attendues')
plt.xlabel('Catégories d\'améliorations')
plt.ylabel('Nombre de réponses')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('reponses_questionnaire.csv')

categories = {
    'Temps de réponse plus rapide': 'Temps de réponse',
    'Meilleure personnalisation des réponses': 'Personnalisation',
    'Meilleure compréhension des besoins des clients': 'Compréhension des besoins',
    'Amélioration des produits/services': 'Amélioration des produits',
    'Aucune amélioration attendue': 'Aucune amélioration'
}

df['Catégorie'] = df['Quelles améliorations espérez-vous de la part des entreprises utilisant l\'analyse des sentiments pour analyser les avis en ligne ?'].map(lambda x: next((cat for key, cat in categories.items() if key in str(x)), 'Autre'))

category_counts = df['Catégorie'].value_counts()
```

```python
#Graphe 1
import seaborn as sns

cross_tab_satisfaction = pd.crosstab(df['Comment évalueriez-vous votre satisfaction globale quant à la manière dont les entreprises gèrent actuellement les avis en ligne ?'], df['Catégorie'])

plt.figure(figsize=(10,6))
sns.heatmap(cross_tab_satisfaction, annot=True, cmap='coolwarm', fmt='d')
plt.title("Corrélation entre satisfaction et catégories d'améliorations")
plt.xlabel("Catégorie d'améliorations")
plt.ylabel("Satisfaction actuelle")
plt.show()
```

```python
#Graphe 2
import seaborn as sns

cross_tab_age = pd.crosstab(df['Âge'], df['Catégorie'])

plt.figure(figsize=(10,6))
sns.heatmap(cross_tab_age, annot=True, cmap='Blues', fmt='d')
plt.title("Catégories d'améliorations en fonction des tranches d'âge")
plt.xlabel("Catégorie d'améliorations")
plt.ylabel("Âge")
plt.show()
```
